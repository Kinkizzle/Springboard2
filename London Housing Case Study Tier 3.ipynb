```python
# A list of Non-Boroughs
non_Boroughs = ['Inner London', 'Outer London', 'NORTH EAST', 'NORTH WEST', 'YORKS & THE HUMBER', 'EAST MIDLANDS', 'WEST MIDLANDS', 'EAST OF ENGLAND', 'LONDON', 'SOUTH EAST', 'SOUTH WEST', 'England']
```

```python
NaNFreeDF2 = NaNFreeDF2[~NaNFreeDF2.London_Borough.isin(non_Boroughs)]
NaNFreeDF2.head()
```

```python
df = NaNFreeDF2
df.head()
df.dtypes
```

```python
camden_prices = df[df['London_Borough'] == 'Camden']
```

```python
ax = camden_prices.plot(kind = 'line', x = 'Month', y = 'Average_price')
```

```python
ax.set_ylabel('Price')
```

```python
df['Year'] = df['Month'].apply(lambda t: t.year)
df.tail()
```

```python
dfg = df.groupby(by=['London_Borough', 'Year']).mean(numeric_only=True)
dfg.sample(10)
```

```python
dfg = dfg.reset_index()
dfg.head()
```

```python
def create_price_ratio(d):
    y1998 = float(d['Average_price'][d['Year'] == 1998])
    y2018 = float(d['Average_price'][d['Year'] == 2018])
    ratio = [y2018/y1998]
    return ratio
```

```python
create_price_ratio(dfg[dfg['London_Borough'] == 'Barking & Dagenham'])
```

```python
final = {}
```

```python
for b in dfg['London_Borough'].unique():
    borough = dfg[dfg['London_Borough'] == b]
    final[b] = create_price_ratio(borough)
print(final)
```

```python
df_ratios = pd.DataFrame(final)
df_ratios.head()
```

```python
df_ratios_T = df_ratios.T
df_ratios = df_ratios_T.reset_index()
df_ratios.head()
```

```python
df_ratios.rename(columns={'index':'Borough', 0:'2018'}, inplace=True)
df_ratios.head()
```

```python
top15 = df_ratios.sort_values(by='2018', ascending=False).head(15)
print(top15)
```

```python
ax = top15[['Borough', '2018']].plot(kind='bar')
ax.set_xticklabels(top15.Borough)
```

```python
# Conclusion: The boroughs depicted in the above graph have seen the greatest increase in housing prices overr the past two decades. In text, above the graph, the top 15 are listed from Hackney (greatest difference) to Greenwich (15th greatest difference)
```
